```python
language_column = df["Мова програмування"]

df_cleaned = df.dropna(axis=1)
df_cleaned["Мова програмування"] = language_column

df_cleaned
```

```python
new_df = df_cleaned.groupby(["Посада"])["Зарплата на місяць"].agg(["min", "max"])

new_df
```

```python
def fill_avg_salary(salary):
  return salary.mean()

new_df["avg"] = new_df.apply(fill_avg_salary, axis=1)
new_df
```

```python
new_df["avg"].describe()
```

```python
new_df.to_csv("/content/drive/My Drive/Colab Notebooks/goit/module_5/Result_2017_jun_final.csv", encoding="utf8")
```

```python
data = {
    "Посада": [
        "Data Scientist / Data Analyst", "DevOps", "ERP / CRM", "HTML coder",
        "Junior Software Engineer", "Product Manager", "Senior Software Engineer",
        "Software Engineer", "System Architect", "Team lead", "Technical Lead"
    ],
    "min": [2700, 7000, 2000, 385, 75, 3000, 350, 100, 400, 2500, 500],
    "max": [2700, 7000, 2000, 600, 2600, 3000, 9200, 7000, 10000, 4000, 18000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

sns.lineplot(x="Посада", y="min", data=df, marker='o', label="Мінімальна зарплата", color="blue")
sns.lineplot(x="Посада", y="max", data=df, marker='o', label="Максимальна зарплата", color="red")


plt.xticks(rotation=90)
plt.xlabel("Посада", fontsize=12)
plt.ylabel("Зарплата", fontsize=12)
plt.title("Мінімальні та максимальні зарплати для різних посад", fontsize=14)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

sns.barplot(x="Посада", y="avg", data=new_df, palette="viridis")

# Оформлення графіка
plt.title("Середня заробітна плата", fontsize=15)
plt.tick_params(labelrotation=90)
plt.ylabel("Середня зарплата", fontsize=12)
plt.xlabel("Посада", fontsize=12)

plt.show()
```
